```
"""
Python code to plot hpc curves
Written by: Kunal Jani
Student ID: 201601444
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
dataset=pd.read_csv('combined_logs.csv')
#print(dataset)
```

```
prob_size=pd.DataFrame(dataset,columns=['Problem Size'])
no_of_threads=pd.DataFrame(dataset,columns=['Threads'])
exec_time=pd.DataFrame(dataset,columns=['Execution time'])
#print(prob_size)
#print(no_of_threads)
#print(exec_time)
```

```
psize=prob_size.drop_duplicates()
thrd=no_of_threads.drop_duplicates()
#print(psize)
#print(thrd)
```

```
ps=psize.iloc[:,0].values
th=thrd.iloc[:,0].values
ext=exec_time.iloc[:,0].values
#print(ps)
#print(th)
#print(ext)
```

```
probs=prob_size.iloc[:,0].values
nthreads=no_of_threads.iloc[:,0].values
#print(probs)
#print(nthreads)
#print(ext)
```

```
no_of_runs=5 #Declare the number of runs here change this value
for i in range(len(th)):
  xtt=[]
  for j in range(len(ps)):
    xt=0
    for k in range(len(ext)):
      if(probs[k]==ps[j] and nthreads[k]==th[i]):
        xt=xt+ext[k]
    xtt.append(xt/(no_of_runs))
  plt.plot(ps,xtt,label=str(th[i])+' threads')
  plt.scatter(ps,xtt)
plt.xlabel('Problem size')
plt.ylabel('Execution time')
plt.title('Execution Time vs Problem Size')
plt.legend()
plt.show()
```
